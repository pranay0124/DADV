```python
import pandas as pd
import numpy as np
```

### 1. Find the 20 most popular movies with a rank more than 8.0

```python
ratings = pd.read_csv("title_ratings.tsv", delimiter = '\t')
```

```python
ratings.head()
```

```python
basics = pd.read_csv("title_basics.tsv", delimiter = '\t')
```

```python
basics.head()
```

```python
finalRatings = ratings[ratings['averageRating'] > 8.0]
finalRatings
```

```python
fr = finalRatings.sort_values('numVotes', ascending = False).head(20)
fr
```

```python
# merging fr and basics dataframes with respect to tconstant
out = pd.merge(fr, basics, on= 'tconst')
out
```

```python
out = out[out['titleType'] == "movie"]
out
```

```python
out['originalTitle'].head(20)
```

### 2. Find the 20 best rated movies with over 40,000 votes in the 2000s (year >= 2000)

```python
type(basics)
```

```python
basics.dtypes
```

```python
basics['startYear'] = pd.to_numeric(basics['startYear'], errors='coerce')
```

```python
basics.dtypes
```

```python
year = basics[basics.startYear >= 2000]
```

```python
year.head()
```

```python
out1 = pd.merge(year, ratings, on= 'tconst')
out1
```

```python
out1 = out1[out1['titleType'] == "movie"]
out1
```

```python
votes = out1[out1.numVotes > 40000]
votes
```

```python
fr1 = votes.sort_values('averageRating', ascending = False).head(20)
fr1
```

```python
print(fr1.originalTitle)
```
