```python
# reading all the csv files in the folder and storing the necessary column in a dictionary
import glob
import csv
import pandas as pd
count = 1
for files in glob.glob('E:\\DADV\\BSE-Assignment\\Datasets\\*.csv'):
    f=open(files, 'r')
#     data = csv.reader(f)
#     contents = f.read()
#     csv_file=csv.writer(f)
#     csv_file.writerow(["gravatar_id","position","number","votes","created_at","comments","body","title","updated_at","html_url","user","labels","state"])
    if not f.read(1):
        dic[str(count)] = [0]
        count = count+1
        continue
    data = pd.read_csv(f)
    dic[str(count)] = list(data["Close Price"])
    count = count + 1
    f.close()
```

### Saving the correlation data in Type1 format

```python
import numpy as np
import math
from scipy import interpolate
corrdic = {}
for each1 in dic.keys():
    lis = []
    for each2 in dic.keys():
#         x = each1
#         y = each2
#         if len(x) < len(y):
#             x = interpolate.interp1d( x, np.linspace( 1, len(x), len(y) ) );
#         else:
#             y = interpolate.interp1d( y, np.linspace( 1, len(y), len(x) ) );
        corr_value = np.corrcoef(dic.get(each1), dic.get(each2)) [0][1]
        if np.isnan(corr_value):
            lis.append(0)
        else:
            lis.append(corr_value)
    corrdic[each1] = lis
```

### Saving the correlation data in Type2 format

```python
import numpy as np
import math
from scipy import interpolate
corrlis = []
for each1 in dic.keys():
    for each2 in dic.keys():
        corr_value = np.corrcoef(dic.get(each1), dic.get(each2)) [0][1]
        if np.isnan(corr_value):
            corrlis.append(0)
        else:
            corrlis.append(corr_value)
```

```python
with open('correlation_data2.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Company1","Company2","Value"])
    i = 1
    for each1 in dic.keys():
        j = 1
        for each2 in dic.keys():
            corr_value = np.corrcoef(dic.get(each1), dic.get(each2)) [0][1]
            if np.isnan(corr_value):
                writer.writerow([i, j, 0])
            else:
                writer.writerow([i, j, corr_value])
            j = j + 1
        i = i+1
```
